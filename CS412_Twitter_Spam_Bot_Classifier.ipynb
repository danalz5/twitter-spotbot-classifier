## Second Step: Cleaning Data

```python
# since I only end up using the text column I drop all NaN texts and drop any duplicate tweets
genuine_tweets = genuine_tweets.dropna(subset=['text']).drop_duplicates(subset=['text'])
social_spambot_tweets = social_spambot_tweets.dropna(subset=['text']).drop_duplicates(subset=['text'])
trad_spambot_tweets = trad_spambot_tweets.dropna(subset=['text']).drop_duplicates(subset=['text'])
```

Im going to get rid of all the rows except text and then create a column for labels.
*   0:   Genuine
*   1:   Spambot



```python
# isolating the text
genuine_text = genuine_tweets[['text']].copy().dropna()
social_spambot_text = social_spambot_tweets[['text']].copy().dropna()
trad_spambot_text = trad_spambot_tweets[['text']].copy().dropna()

# keeping 50k genuine tweets, 25k social spambot tweets, and 25k traditional spambot tweets
# total number of tweets 100k with a 50/50 balance
genuine_text = genuine_text.sample(50000)
social_spambot_text = social_spambot_text.sample(25000)
trad_spambot_text = trad_spambot_text.sample(25000)

# 20k that weren't sampled is going to be used at the validation set
genuine_val = genuine_tweets[['text']].drop(genuine_text.index).dropna().sample(10000)
social_spambot_val = social_spambot_tweets[['text']].drop(social_spambot_text.index).dropna().sample(5000)
trad_spambot_val = trad_spambot_tweets[['text']].drop(trad_spambot_text.index).dropna().sample(5000)

# 20k of what wasn't sampled for either val or train saved for test
genuine_test = genuine_tweets[['text']].drop(genuine_text.index).drop(genuine_val.index).dropna().sample(10000)
social_spambot_test = social_spambot_tweets[['text']].drop(social_spambot_text.index).drop(social_spambot_val.index).dropna().sample(5000)
trad_spambot_test = trad_spambot_tweets[['text']].drop(trad_spambot_text.index).drop(trad_spambot_val.index).dropna().sample(5000)

genuine_text['label'] = 0 # genuine label = 0
social_spambot_text['label'] = 1 # social spambot = 0
trad_spambot_text['label'] = 1 # traditional spambot = 0

genuine_val['label'] = 0
social_spambot_val['label'] = 1
trad_spambot_val['label'] = 1

genuine_test['label'] = 0
social_spambot_test['label'] = 1
trad_spambot_test['label'] = 1

# now I'm gonna combine and shuffle these seperate data frames to create
# a training, a validation set, and a testing set
train_set = pd.concat([genuine_text, social_spambot_text, trad_spambot_text]).sample(frac=1).reset_index(drop=True)
val_set = pd.concat([genuine_val, social_spambot_val, trad_spambot_val]).sample(frac=1).reset_index(drop=True)
test_set = pd.concat([genuine_test, social_spambot_test, trad_spambot_test]).sample(frac=1).reset_index(drop=True)

# orignally high accuracy, precision, recall, and f1 scores game from huge overlap in the same tweets
# making sure there is minumum overlap
overlap = pd.merge(train_set[['text']], val_set[['text']], on='text', how='inner')
print("Number of overlapping tweets:", len(overlap))
```

## Third Step: Establishing a Logistic Regression Baseline

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
import time

best_accuracy = 0
best_hyperparam = None
hyperparams = [64, 128, 256, 512, 1024]

start = time.time()
for num_features in hyperparams:
  vectorizer = TfidfVectorizer(max_features=num_features)
  training_vector = vectorizer.fit_transform(train_set['text']) # learn the TF-IDF
  val_vector = vectorizer.transform(val_set['text']) # create vectors based on learned TF-IDF

  logistic_regression_classifier = LogisticRegression(max_iter=1000) # initialize the model
  logistic_regression_classifier.fit(training_vector, train_set['label']) # feeding the mode training vectors and their corresponding labesl

  val_outputs = logistic_regression_classifier.predict(val_vector) # predict on the validation set
  print(f"Accuracy: {accuracy_score(val_set['label'], val_outputs)} for {num_features} features")
  if accuracy_score(val_set['label'], val_outputs) > best_accuracy:
    # calculate metrics for the best found version of the model
    best_hyperparam = num_features
    best_accuracy = accuracy_score(val_set['label'], val_outputs)
    best_precision = precision_score(val_set['label'], val_outputs)
    best_recall = recall_score(val_set['label'], val_outputs)
    best_f1 = f1_score(val_set['label'], val_outputs)
end = time.time()
print(f"Hyperparameter training took {end-start:.1f} seconds.")

# after finding the best number of features, find the metrics for the testing set
start = time.time()
vectorizer = TfidfVectorizer(max_features=best_hyperparam)
training_vector = vectorizer.fit_transform(train_set['text'])
test_vector = vectorizer.transform(test_set['text'])
logistic_regression_classifier = LogisticRegression(max_iter=1000)
logistic_regression_classifier.fit(training_vector, train_set['label'])
test_outputs = logistic_regression_classifier.predict(test_vector)
print(f"Best number of features: {best_hyperparam}")
print(f"\nMetrics on the test set: {best_hyperparam}")
print(f"Accuracy  : {accuracy_score(test_set['label'], test_outputs):.3f}")
print(f"Precision : {precision_score(test_set['label'], test_outputs):.3f}")
print(f"Recall    : {recall_score(test_set['label'], test_outputs):.3f}")
print(f"F1        : {f1_score(test_set['label'], test_outputs):.3f}")
end = time.time()
print(f"\nThe best model was trained and tested in {end-start:.1f} seconds.")
```

##### Prepping The Data For The BERT Model


1.   Converting the pandas dataframe to a hugging face Dataset
2.   Creating a tokenizer that uses distilibert
3.   Tokenizing the training and validation datasets


```python
from transformers import AutoTokenizer
from datasets import Dataset

BERT_model_name = "huawei-noah/TinyBERT_General_4L_312D"
tokenizer = AutoTokenizer.from_pretrained(BERT_model_name)
# changing from dataframe to dataset to be more compatiable with BERT, and have useful functions like .map()
bert_train_set = Dataset.from_pandas(train_set)
bert_val_set = Dataset.from_pandas(val_set)
bert_test_set = Dataset.from_pandas(test_set)
def create_tweet_tokens(tweets): # tokenizer function to pass into Dataset.map()
  # tweets that are too long will be truncated to 64 tokens
  # 64 so there's not too much padding for shorter tweets
  return tokenizer(tweets['text'], max_length=64, truncation=True, padding="max_length")

# tokenizer the train and val set using the distilibert tokenizer
bert_train_set = bert_train_set.map(create_tweet_tokens, batched=True)
bert_val_set = bert_val_set.map(create_tweet_tokens, batched=True)
bert_test_set = bert_test_set.map(create_tweet_tokens, batched=True)
```

##### Setting Up The BERT Model For Finetuning
This took about an hour so don't run it again, the best model is saved to google drive

```python
def get_metrics (test_labels): # metric function to pass into model for evaluation
  logits, actual_label = test_labels
  test_output = logits.argmax(axis=-1) # get the predicted class from the output
  return {'accuracy': accuracy_score(actual_label, test_output), 'precision': precision_score(actual_label, test_output), 'recall': recall_score(actual_label, test_output), 'f1': f1_score(actual_label, test_output)}
```

```python
from transformers import TrainingArguments, Trainer, AutoModelForSequenceClassification
import os

learning_rates = [0.0001, 0.0002, 0.0004]
batches = [128, 64] # anything lower than 64 takes too long to train
best_accuracy = 0
best_hyperparams = {}
best_model = None
hyper_param_tune = False # this is false so i dont accidently start the hyperparam tuning again because it takes forever, set this to true to find optimal hyper params
if hyper_param_tune:
  for rate in learning_rates: # grid search for best hyperparam combo
    for batch in batches:
        model = AutoModelForSequenceClassification.from_pretrained('huawei-noah/TinyBERT_General_4L_312D',  num_labels=2) # load pretrained tinyBERT model
        training_hyperparams = TrainingArguments(
          output_dir = path + "results",
          learning_rate = rate,
          per_device_train_batch_size = batch,
          per_device_eval_batch_size = batch,
          num_train_epochs = 4, # anything more than 4 epochs leads to >10 mins per train
          report_to = [], # so it doesn't request an API key
          fp16 = True # 16-bit floating point for faster training
        )

        model_trainer = Trainer( # initialize a trainer specifying mode, hyperparams, and dataset as well as custom eval function
          model = model,
          args = training_hyperparams,
          train_dataset = bert_train_set,
          eval_dataset = bert_val_set,
          compute_metrics = get_metrics
        )

        model_trainer.train()
        metrics = model_trainer.evaluate()

        accuracy = metrics.get("eval_accuracy")
        print(accuracy)
        if accuracy > best_accuracy: # find the best hyperparams and save the best model
          best_accuracy = accuracy
          best_hyperparams = {"learning_rate": rate, "batch_size": batch}
          best_model = model

  print("Best params:", best_hyperparams)
  print("Best accuracy:", best_accuracy)
  save_dir = '/content/drive/MyDrive/CS412_spambot_classifier/best_model/' # save the best model to my google drive
  os.makedirs(save_dir, exist_ok=True)
  best_model.save_pretrained(save_dir)
```

Using the best found hyperparameters from above to train and log training set and validation set losses.

```python
%load_ext tensorboard
model = AutoModelForSequenceClassification.from_pretrained('huawei-noah/TinyBERT_General_4L_312D',  num_labels=2) # load pretrained tinyBERT model
training_hyperparams = TrainingArguments(
  output_dir = path + "results",
  logging_dir = path + "logs",
  logging_steps = 100,
  learning_rate = 0.0001, # best found learning rate
  per_device_train_batch_size = 64, # best found batch_size
  per_device_eval_batch_size = 64,
  num_train_epochs = 4, # anything more than 4 epochs leads to >10 mins per train
  report_to = ['tensorboard'], # so it doesn't request an API key
  fp16 = True, # 16-bit floating point for faster training
)

model_trainer = Trainer(
  model = model,
  args = training_hyperparams,
  train_dataset = bert_train_set,
  eval_dataset = bert_val_set,
  compute_metrics = get_metrics
)

model_trainer.train()
```

Final metrics for the fine-tuned TinyBERT model!

```python
best_model = AutoModelForSequenceClassification.from_pretrained('/content/drive/MyDrive/CS412_spambot_classifier/best_model/')
eval_args = TrainingArguments(
  output_dir="./tmp",
  per_device_eval_batch_size=256,
  do_train=False, # the model is already trained
  report_to=[], # set so I'm not ask for an api key
  fp16=True # 16bit float point precision for faster train/eval/test
)

eval_trainer = Trainer( # final evalution on the test set, loading in the best model found during hyperparamter finetuning
  model=best_model,
  args=eval_args,
  eval_dataset=bert_test_set,
  compute_metrics=get_metrics
)

metrics = eval_trainer.evaluate()
print(f"\ntinyBERT metrics on the test set:")
print(f"Accuracy  : {metrics['eval_accuracy']:.3f}")
print(f"Precision : {metrics['eval_precision']:.3f}")
print(f"Recall    : {metrics['eval_recall']:.3f}")
print(f"F1        : {metrics['eval_f1']:.3f}")
```
